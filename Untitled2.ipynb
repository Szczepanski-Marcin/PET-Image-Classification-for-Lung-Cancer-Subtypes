```python
import os
import cv2
import pydicom
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# ==============================
# CONFIG
# ==============================
DATASET_PATH = "/Users/Marcin_1/Desktop/PET/imbalanced_dataset"
VIDEO_OUTPUT = "/Users/Marcin_1/Desktop/PET"
FPS = 10

# ==============================
# LOAD DICOM FILES
# ==============================
def load_dicom_files(folder):
    files = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith('.dcm')]

    def sort_key(f):
        d = pydicom.dcmread(f)
        return getattr(d, "InstanceNumber", 0)

    files.sort(key=sort_key)
    return files

# ==============================
# FEATURE EXTRACTION
# ==============================
def extract_features(img):
    roi_size = np.sum(img > np.mean(img))
    mean_intensity = np.mean(img)
    max_intensity = np.max(img)
    return roi_size, mean_intensity, max_intensity

# ==============================
# BUILD DATASET (SIMPLE LABELING)
# ==============================
def build_dataset(files):
    rows = []

    for file in files:
        try:
            dcm = pydicom.dcmread(file)
            img = dcm.pixel_array.astype(float)

            # Normalize
            img = (img - np.min(img)) / (np.max(img) - np.min(img))

            roi, mean_i, max_i = extract_features(img)

            # Dummy label (replace with real labels if available)
            label = 1 if max_i > 0.8 else 0

            rows.append({
                "ROI_Size": roi,
                "Mean_Intensity": mean_i,
                "Max_Intensity": max_i,
                "Label": label
            })
        except Exception as e:
            print("Error:", e)

    df = pd.DataFrame(rows)
    return df

# ==============================
# TRAIN MODEL
# ==============================
def train_model(df):
    if len(df) == 0:
        raise ValueError("Dataset is empty")

    X = df[["ROI_Size", "Mean_Intensity", "Max_Intensity"]]
    y = df["Label"]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    model = RandomForestClassifier()
    model.fit(X_train, y_train)

    acc = model.score(X_test, y_test)
    print(f"Model accuracy: {acc:.2f}")

    return model

# ==============================
# PROCESS & CREATE VIDEO
# ==============================
def create_video(files, model):
    frames = []

    for i, file in enumerate(files):
        dcm = pydicom.dcmread(file)
        img = dcm.pixel_array.astype(float)

        # Normalize
        img = (img - np.min(img)) / (np.max(img) - np.min(img))
        img_uint8 = (img * 255).astype(np.uint8)

        # Feature extraction
        roi, mean_i, max_i = extract_features(img)
        features = np.array([[roi, mean_i, max_i]])

        # Prediction
        pred = model.predict(features)[0]
        prob = model.predict_proba(features)[0][pred]

        # Threshold detection
        _, thresh = cv2.threshold(img_uint8, 200, 255, cv2.THRESH_BINARY)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        img_color = cv2.applyColorMap(img_uint8, cv2.COLORMAP_JET)

        for cnt in contours:
            if cv2.contourArea(cnt) > 50:
                x, y, w, h = cv2.boundingRect(cnt)
                cv2.rectangle(img_color, (x, y), (x+w, y+h), (0,255,0), 2)

        # Overlay text
        label_text = "Cancer" if pred == 1 else "Normal"
        cv2.putText(img_color, f"Prediction: {label_text}", (10,30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
        cv2.putText(img_color, f"Confidence: {prob:.2f}", (10,60),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
        cv2.putText(img_color, f"Slice: {i}", (10,90),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)

        frames.append(img_color)

    h, w, _ = frames[0].shape
    out = cv2.VideoWriter(VIDEO_OUTPUT, cv2.VideoWriter_fourcc(*'mp4v'), FPS, (w, h))

    for f in frames:
        out.write(f)

    out.release()
    print("Video saved:", VIDEO_OUTPUT)

# ==============================
# MAIN PIPELINE
# ==============================
def main():
    files = load_dicom_files(DATASET_PATH)
    print("Loaded files:", len(files))

    df = build_dataset(files)
    print("Dataset size:", len(df))

    model = train_model(df)
    create_video(files, model)

if __name__ == "__main__":
    main()
```
